Rede Perceptron para o porblema dos Fil√≥sofos vs Compositores

O problema consiste em classificar as 4 entradas de acordo com a seguinte tabela
image_2022-11-28_183709251.png



```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.activations import hard_sigmoid, sigmoid, softmax
```

```python
X = [[0,0],[0,1],[1,0],[1,1]]
Y = [[0],[1],[1],[0]]
```

```python

model = Sequential()
model.add(Dense(8, input_shape=(2, ), activation=sigmoid, use_bias=True))
model.add(Dense(1, activation=hard_sigmoid))

opt = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer=opt)

model.fit(
    X,
    Y,
    epochs=1000,
    batch_size=1,
    verbose=1,
    shuffle=True
)

_, accuracy = model.evaluate(X, Y)
print("%0.3f" % accuracy)
```

```python
model.predict([[0,0], [0, 1], [1, 0], [1, 1]])
```
