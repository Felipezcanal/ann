Rede Perceptron para o porblema dos Fil√≥sofos vs Compositores

O problema consiste em classificar as 4 entradas de acordo com a seguinte tabela
image_2022-11-28_183709251.png



```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.activations import hard_sigmoid
```

```python
X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]     
]
Y = [
    [0],
    [0],
    [1],
    [1]
]
```

```python

model = Sequential()
model.add(Dense(1, input_shape=(2,), kernel_initializer=tf.initializers.Ones, activation=hard_sigmoid, use_bias=True, bias_initializer='zeros'))

opt = tf.keras.optimizers.SGD(learning_rate=1)
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer=opt)

model.fit(X, Y, epochs=10, batch_size=25, verbose=1, shuffle=True)

_, accuracy = model.evaluate(X, Y)
print("%0.3f" % accuracy)
```

```python
model.predict([[0,0], [0, 1], [1, 0], [1, 1]])
```
